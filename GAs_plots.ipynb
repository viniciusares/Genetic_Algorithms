```python
# Genetic Algorithm for approximating linear function
# Vinicius Ares - 2018/10/06
# This code plots points and lines
import random
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# Plotting points P1 and P2
P1 = [2, 3]
P2 = [6, 5]
p1p2_Xs = [P1[0], P2[0]] # contains the X values of P1 and P2
p1p2_Ys = [P1[1], P2[1]] # contains the Y values of P1 and P2
plt.plot(p1p2_Xs, p1p2_Ys, 'ro')

# Creates and plots random line equations
xvecplot = np.array([-6, 6]) # It's a vector (linear algebra)
yvecplot = []
for line in range(10):
    a = random.uniform(-1,1) # Slope
    b = random.uniform(-10,10) # Intersect
    # Recieves 2 Xs and outputs 2 Ys of a line equation
    yvecplot.append(a*xvecplot + b)
    plt.plot(xvecplot, yvecplot[line])
```
