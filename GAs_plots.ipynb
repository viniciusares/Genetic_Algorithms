```python
# Genetic Algorithm for approximating linear function
# Vinicius Ares - 2018/10/06
# This code plots points and lines and prints errors relative to points, unsorted and sorted
import random
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

def round_list2D(list_of_lists, decimals=3):
    """Round a list of lists to some given number of decimals"""
    rounded_list = [[round(n2, decimals) for n2 in list_of_lists[n1]] for n1 in range(len(list_of_lists))]
    return rounded_list

def TakeFirst(elem):
    return elem[0]

# Plotting points P1 and P2
P1 = [2, 3]
P2 = [6, 5]
p1p2_Xs = [P1[0], P2[0]] # contains the X values of P1 and P2
p1p2_Ys = [P1[1], P2[1]] # contains the Y values of P1 and P2
plt.plot(p1p2_Xs, p1p2_Ys, 'ro')

# Creates and plots random line equations, and calculate errors
xvecplot = np.array([-6, 6]) # It's a vector (linear algebra)
yvecplot, list_er_all, list_er_a_b = [], [], []
for individual in range(10):
    a = random.uniform(-1,1) # Slope
    b = random.uniform(-10,10) # Intersect
    # Recieves 2 Xs and outputs 2 Ys of a line equation
    yvecplot.append(a*xvecplot + b)
    plt.plot(xvecplot, yvecplot[individual])
    Er_toPt1 = a*P1[0]+b - P1[1] # output - desired
    Er_toPt2 = a*P2[0]+b - P2[1] # output - desired
    er_individual = (Er_toPt1**2 + Er_toPt2**2)**0.5
    list_er_a_b.append([er_individual, a, b])

#plt.show() # not necessary in jupyter notebook

print('[er,a,b], before sort:\n', round_list2D(list_er_a_b, 2), '\n')

list_er_a_b.sort(key=TakeFirst)
print('Sorted [er,a,b]:\n', round_list2D(list_er_a_b, 2), '\n')
```
